```python
rows = []
for i in range(1, 16):
    print (i)
    page = urllib.request.urlopen("https://eid2.liverpool.ac.uk/OrganismCountryInteractions?Grid-page=%i&Grid-orderBy=~&Grid-filter=Class~eq~'viruses'~and~NoSequences~ge~100" % i)
    soup = BeautifulSoup(page)  
    s = soup.findAll('tr')
    for v in s[1:]:
        tags = v.findAll(text=True)[0:7]
        rows.append(tags)
```

```python
country_tab = pd.DataFrame(rows)
```

```python
country_tab = country_tab.drop(5, 1)
```

```python
country_tab = country_tab.drop(3, 1)
```

```python
country_tab.columns = ['virus', 'virus_taxid', 'virus_rank', 'country', 'nrecords']
```
